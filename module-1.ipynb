## 2. Model Overview

A general overview of the Address Model content - how it works and what it depends on.

_This section is based on the model as presented at <https://linked.data.gov.au/def/addr>._

### What

<img src="module-1/img/model-overview.svg" style="width:750px;" />

#### Addresses

* **Addresses** are special kinds of names for features, made of Parts
* Addresses parts are either simple values - street number - or complex objects - Locality
* Complex object parts can be made of other parts - "Smith Street" made of "Smith" & "Street"
* Parts are assigned roles/types for ordering & templating
* Parts can vary between model uses without changing the model

#### Geocodes

* **Geocodes** are geometries with persistent roles - `Property Access Point`

#### Addressable Objects

* Addresses are assigned to **Addressable Objects**
* Addressable Objects can be anything a user allows to have an address assigned to it
    * in Qld: all Parcels, many Properties, aliases

### Why

* Meets all the _Strategy Technical Addendum_'s "Model Directions"
* Reuses existing models (G-NAF, ISO, AS)
* Reuses standardised fundamental model patterns - feature naming, lifecycling, spatiality, provenance
* Allows for customisation in vocabs and _profiles_ - additional constraints
* Is formulated as part of a model platform with Geographical (Place) Names, Roads & Cadastre

### How

<img src="module-1/img/iso19160.png" style="width:500px; float:right;" />

* emulates [ISO 19160](https://www.iso.org/standard/61710.html), international addressing standard
    * can be exported to match
    * ISO 19160 2027/2028 will adopt this model
* follows same Name/Feature pattern used for Geographical/Place Names & Road Names
* works with [GN](https://linked.data.gov.au/def/gn), [Road Names](https://linked.data.gov.au/def/roads), [CSDM](https://linked.data.gov.au/def/csdm) and [CAD](https://linked.data.gov.au/def/cad) models
    * by emulating Feature Name / Feature pattern
    * by reussing background patterns    
* relies on published vocabs
    * e.g. [Address Part Types](https://linked.data.gov.au/def/addr-part-types)
* provides multiple resource:
    * conceptual model
    * validator
    * vocabularies

### Who

* Model developed by Nicholas Car for Qld then ICSM, 2020 - 2025
    * based on ISO 19160 & GNAF modelling by N. Car & PSMA, 2016+
    * with contributions from Qld Spatial Information & Edmond Chuc
    * incorporates modelling by N. Car for Geoscience Australia - Place Names
    * reviewed by Address Working Group (AWG) for adoption in 2023
* Governed by the AWG

### When

* initially developed, 2020 - 2022
* issued in 2023
* polished 2023+

### 3.4.1 Conceptual Model

* presented as a UML Class Diagram-like model, as per W3C standards
* a [formal OWL diagram](https://www.semantic-web-journal.net/content/g-owl-complete-visual-syntax-owl-2-modeling-and-communication) can be made from schema, if needed
* Classes and Properties defined in English
* Example for `Address` class given in a graph data syntax ("turtle"):

```turtle
# 20 Oxford Place, Shorncliffe QLD 4017
#
PREFIX addr: <https://linked.data.gov.au/def/addr/>
PREFIX apt: <https://linked.data.gov.au/def/addr-part-types/>
PREFIX cn: <https://linked.data.gov.au/def/cn/>
PREFIX ex: <http://example.com/>
PREFIX ex-parcel: <http://example.com/parcel/>
PREFIX geo: <http://www.opengis.net/ont/geosparql#>
PREFIX ls: <https://linked.data.gov.au/def/lifecycle/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX schema: <https://schema.org/>

<http://example.com/oxford>
    a addr:Address ;
    schema:hasPart [
        schema:additionalType apt:numberFirst ;
        schema:value 20
    ] ,
    [
        schema:additionalType apt:road ;
        schema:value <https://linked.data.gov.au/dataset/qldaddr/streetLocality/QLD140492> ; # "Oxford Place"
    ] ,
    [
        schema:additionalType apt:locality ;
        schema:value <https://linked.data.gov.au/dataset/qldaddr/locality/loc38f189794e03> ;  # "Shorncliffe"
    ] ,
    [
        schema:additionalType apt:stateOrTerritory ;
        schema:value <https://linked.data.gov.au/dataset/asgsed3/STE/3> ; # "Queensland"
    ] ,
    [
        schema:additionalType apt:postcode ;
        schema:value 4017
    ] ,
    [
        schema:additionalType apt:country ;
        schema:value <https://linked.data.gov.au/dataset/asgsed3/AUS/AUS> ; # "Australia"
    ] ;
    addr:hasGeocode [
        schema:roleName <https://linked.data.gov.au/def/geocode-types/property-centroid> ;
        schema:value [
            geo:asWKT "POINT(153.0846 -27.3270)"^^geo:wktLiteral ;
        ] ;
    ] ;
    cn:isNameFor ex-parcel:161162441 ;
    ls:hasLifecycleStage [
        schema:additionalType ex:Unofficial ;
        schema:startTime "1932"^^xsd:date ;
    ] ;
.
```

### 3.4.3 Vocabularies

The 'slots' in the model that need specific values - address part types, Geocode role etc. - are to be drawn from the Address vocabularies on the ICSM vocabulary server: <https://icsm.information.qld.gov.au>.

#### Address Part Types

<https://linked.data.gov.au/def/addr-part-types>

Values such as [Address Number First](https://linked.data.gov.au/def/addr-part-types/addressNumberFirst), [Locality](https://linked.data.gov.au/def/addr-part-types/locality), [Indigenous Country Name](https://linked.data.gov.au/def/addr-part-types/indigenousCountryName).

This vocabulary is expandable and can be subset by jurisdictions.

Management of the vocabularies is implemented according to ISO 19135 - _Procedures for item registration_ - but is only now becoming 'live' nationally as multiple jurisdictions are carefully examining the values.

### 3.5.2 Templating

As per the [Address Model's Annex C: Templates](https://linked.data.gov.au/def/addr#_basic_template), we can render address information according to different templates, e.g.:

Long form:
```
The Manse
Unit 4B
72 Yundah Street
Shorncliffe, Queensland
4017 Australia
```

Abbreviated:
```
4B/72 Yundah St
Shorncliffe, QLD
4017 Aust.
```

Single line:
```
4B/72 Yundah St Shorncliffe QLD 4017
```

A formal templating grammar has been specified in the Model and some alternate form logic - "Level 8" &rarr; "Lv. 8" - but specific implementations are left to jurisdictions.
